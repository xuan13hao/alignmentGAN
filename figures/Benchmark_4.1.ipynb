```python
aver_pattern = r'oracle_sample_NLL = (.*)'
aver_file = "log_04_01.txt"
oracle_NLL = []
eps = []
k = 0
with open(aver_file, 'r') as f:
    for line in f:
        # print(line)
        result = re.search(aver_pattern, line)
        # print(result)
        if result:
            k = k + 1
        # print(result)
            average_train_NLL = float(result.group(1))
            oracle_NLL.append(average_train_NLL)
            eps.append(k)
```
