```python
import torch
import re
import matplotlib.pyplot as plt
file_path = 'pre-generator_3.16.txt'
i = 0
ave_train_null = []
ora_train_null = []
epochs = []
pattern = r"average_train_NLL = (\d+\.\d+), oracle_sample_NLL = (\d+\.\d+)"

with open(file_path, 'r') as f:
    for line in f:
        i = i + 1
        result = re.search(pattern, line)
        if result:
            average_train_NLL = float(result.group(1))
            oracle_sample_NLL = float(result.group(2))
            ave_train_null.append(average_train_NLL)
            ora_train_null.append(oracle_sample_NLL)
            epochs.append(i)
            #print("average_train_NLL = {}, oracle_sample_NLL = {}".format(average_train_NLL, oracle_sample_NLL))
        else:
            print("No match found.")
```

```python
plt.plot(epochs, ave_train_null)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Pre-train Generator Curve')
plt.show()
```

```python
plt.plot(epochs, ora_train_null)
plt.xlabel('Epoch')
plt.ylabel('Oracle Train Null')
plt.title('Oracle Train Curve')
plt.show()
```

```python
dis_file = "pre-discrinimator_3_16.txt"
dis_pattern = r'd-step\s\d+\sepoch\s\d+\s:\s\.+\saverage_loss\s=\s(\d+\.\d+),\strain_acc\s=\s(\d+\.\d+),\sval_acc\s=\s(\d+\.\d+)'
ave_los = []
train_los = []
ep = []
j = 0
with open(dis_file, 'r') as f:
    for line in f:
        j = j + 1
        # print(line)
        result = re.search(dis_pattern, line)
        if result:
        # print(result)
            average_train_NLL = float(result.group(1))
            oracle_sample_NLL = float(result.group(2))
        # print(average_train_NLL)
            ave_los.append(average_train_NLL)
            train_los.append(oracle_sample_NLL)
            ep.append(j)
        # print(epochs)
            #print("average_train_NLL = {}, oracle_sample_NLL = {}".format(average_train_NLL, oracle_sample_NLL))
```

```python
plt.plot(ep, ave_los)
plt.xlabel('Epoch')
plt.ylabel('Average Loss')
plt.title('Discrinimator Train Curve')
plt.show()
```

```python
plt.plot(ep, train_los)
plt.xlabel('Epoch')
plt.ylabel('Train Average Accuracy')
plt.title('Discrinimator Average Curve')
plt.show()
```

```python
aver_pattern = r'oracle_sample_NLL = (.*)'
aver_file = "log_3_16.txt"
oracle_NLL = []
eps = []
k = 0
with open(aver_file, 'r') as f:
    for line in f:
        # print(line)
        result = re.search(aver_pattern, line)
        # print(result)
        if result:
            k = k + 1
        # print(result)
            average_train_NLL = float(result.group(1))
            oracle_NLL.append(average_train_NLL)
            eps.append(k)
```

```python
plt.plot(eps, oracle_NLL)
plt.xlabel('Epoch')
plt.ylabel('Oracle Loss')
plt.title('Oracle Curve')
plt.show()
```
